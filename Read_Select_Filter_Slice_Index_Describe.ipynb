## Indexing

#### Using slicing to make changes is not recommended as it changes original dataframe directly
#### Thus we use loc and iloc methods of pandas for selecting rows and columns if any column or row needs changes

```python
#Notice unlike standard python slicing, end index 5 is inclusive in output
df.loc[2:5, 'title']
```

```python
#Let's change these values of lower case
df.loc[2:5, 'title'] = df.loc[2:5, 'title'].str.lower()
df.loc[2:5, 'title']
```

```python
#Select multiple columns
df.loc[2:5, ['title', 'mag']]
```

```python
#iloc can be used to access the dataframe with integer location 
#it works similar to python slicing i.e. end index is excluded
df.iloc[2:5, [19, 8]]
```

```python
#Selecting range columns like rows
df.iloc[2:5, 2:5]
```

```python
#access value of specific row and column name
df.at[10, 'title']
```

```python
#access value based on location
df.iat[10, 19]
```

# Filter

```python
#Comparison
df.mag > 5
```

```python
#Display the df based on condition
#Only true rows will be returned
df[df.mag > 5]
```

```python
#Display specific columns based of condition
df.loc[df.mag>=7, ['alert', 'mag', 'magType', 'title', 'tsunami', 'type']]
```

```python
#Multiple conditions
#Satisfies both conditions
df.loc[(df.tsunami==1) & (df.alert=='red'), ['alert', 'mag', 'magType', 'title', 'tsunami', 'type']]
```

```python
#Multiple conditions
#Satisfies at least of the conditions
df.loc[(df.tsunami==1) | (df.alert=='red'), ['alert', 'mag', 'magType', 'title', 'tsunami', 'type']]
```

```python
df.loc[(df.place.str.contains('Alaska')) & (df.alert.notnull()), ['alert', 'mag', 'magType', 'title', 'tsunami', 'type']]
```

```python
#Using regex
df.loc[(df.place.str.contains(r'CA|California$')) & (df.mag>=3), ['alert', 'mag', 'magType', 'title', 'tsunami', 'type']]
```

```python
#Conditin based on range of values
#Both ranges inclusive
df.loc[(df.mag.between(5,8)) & (df.alert.notnull()), ['alert', 'mag', 'magType', 'title', 'tsunami', 'type']].head(3)
```

```python
#Condition mathing with specific values 
df.loc[df.alert.isin(['orange', 'red']), ['alert', 'mag', 'magType', 'title', 'tsunami', 'type']]
```

```python
#Getting the row number of max and min values of specific column
print(df.mag.idxmax(), df.mag.idxmin())
```
